```python
import matplotlib.pyplot as plt
import numpy as np
import netCDF4 as nc
```

```python
# Open the netCDF file
nc_file = '/content/gebco_2025_n-8.0049_s-8.7932_w122.6074_e123.1897.nc'
dataset = Dataset(nc_file)
```

```python
# Read the necessary variables
lon = dataset.variables['lon'][:]
lat = dataset.variables['lat'][:]
bathymetry = dataset.variables['elevation'][:]

sea = np.where(bathymetry < 0, bathymetry, np.nan)
land = np.where(bathymetry > 0, bathymetry, np.nan)
dataset.close()
```

```python
# Create a meshgrid for lon and lat
lon_mesh, lat_mesh = np.meshgrid(lon, lat)
```

```python
# Create the contour plot
plt.figure(figsize=(10, 8))
plt.contourf(lon_mesh, lat_mesh, land, levels=20, cmap='terrain')
plt.contour(lon_mesh, lat_mesh, sea, levels=10, cmap='jet')
plt.colorbar(label='Bathymetry (m)')
plt.title('GEBCO Bathymetry')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
